```python
from transformers import pipeline

# Download model
# https://dataloop.ai/library/model/facebook_bart-large-mnli/
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
text = "I am looking for a flat white"
labels = ['education', 'politics', 'technology', 'cooking']

result = classifier(text, labels)

print(f"Text: {result['sequence']}\n")
for label, score in zip(result['labels'], result['scores']):
    print(f"Label: {label} | Confidence: {score:.2%}")
```

### Multilabel Classification

```python
text = "Angela Merkel spoke at a school event"
labels = ['education', 'politics', 'technology', 'cooking']

result = classifier(text, labels, multi_label=True)

print(f"Text: {result['sequence']}\n")
for label, score in zip(result['labels'], result['scores']):
    print(f"Label: {label} | Confidence: {score:.2%}")
```
